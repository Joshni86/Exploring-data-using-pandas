```python
import pandas as pd
```

```python
df=pd.read_csv('Pokemon.csv')
print(df.head(10))
print(df.tail(10))
```

```python
print("Reading just the headers")
# Read titles as list
print(df.columns.tolist())

print("Reading top 5 rows")
# Read the name and types of top 5 rows
print(df[['Name','Type 1','Type 2']][0:5])

print("Reading first row");
#Read row based on index -- integer location
print(df.iloc[1])

print("Reading a few rows together using iloc")
print(df.iloc[29:37])

#Reading a specifc location -- like specific cell
print("Reading the first column of 2 indexed row")
print(df.iloc[2,1])

# Getting location with a specific condtion:
df.loc[df['Legendary']==True]
```

```python
print("ITERATION")
#Iterating thru rows
print("Iterating thruuuu")
for index,row in df.iterrows():
    print(index,row)
print("Iterating thru just the name column")
for index,row in df.iterrows():
    print(index,row['Name'])
```

```python
print("Describing Data")
df.describe()
```

```python
print("Sort with names")
df.sort_values('Name')

print("Sort in descending order")
df.sort_values('Name',ascending=False)

print("Sorting using multiple columns -- one column asc and one column desc")
df.sort_values(['Name','Type 1'],ascending=[True,False])
```
